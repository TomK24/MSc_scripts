```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import numpy as np
```

```python
df = pandas.read_csv("output_sorted_2.csv", sep=',', header=0, index_col=0) #import csv
#df = df.T
#df.columns += 1 
plt.figure(figsize=(8, 10))
arr = df.to_numpy() #Need to convert dataframe to numpy array to work with np.where() method I think 
sns.set(font_scale=1)
sns.set_style("ticks")
print(df)
#print(arr)
mask = np.where(arr > 0, False, True) # Making a mask array so zero values are not plotted. If arr[i,j] > 0 return False, if not return True. This is then passed to the heatmap. Doing this so the figure isnt filled with zeros in all the cells where there is not hydrogen bond occupancy
#print(mask)
#mask[np.triu_indices_from(mask)] = True
heat_map = sns.heatmap(df, cmap="Blues",vmax=100, vmin=0,yticklabels=1,xticklabels=1, annot=True, mask=mask, fmt='.0f', cbar_kws={"aspect":25, "pad": 0.01}) #plot the heatmap mask= df> 0,  yticklabels="auto",
#mask = np.triu(np.ones_like(df, dtype=np.bool))
heat_map.set_yticklabels(heat_map.get_yticklabels(),rotation=360,fontdict={"fontsize":12})
heat_map.set_xticklabels(heat_map.get_xticklabels(), fontdict={"fontsize":12}, rotation=90)

heat_map.set_xlabel("Residue",fontsize=15)
heat_map.set_ylabel("Ligand",fontsize=15)
# heat_map.set_title("PZA",fontsize=25)
#plt.show()
plt.tight_layout()
plt.savefig('testing.png', format='png', dpi=300, bbox_inches='tight')
```

```python
import pandas

import seaborn as sns
import matplotlib.pyplot as plt


df = pandas.read_csv("Sanc00467_chain_b.csv", sep=',') #import csv

df = df.T
df.columns += 1 

sns.set_style("ticks")
sns.set(font_scale=1.9)

g = sns.clustermap(df,row_cluster=True, col_cluster=False, cmap = "gist_rainbow", vmax=1.0, vmin=0.0, yticklabels=True, figsize=(20, 25))
g.ax_row_dendrogram.set_visible(False)
    
plt.tight_layout()

plt.savefig('sanc00467_rmsf_chain_b_new.png', format='png', dpi=300, bbox_inches='tight')
```
